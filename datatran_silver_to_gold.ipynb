```sql
-- CREAT DIM TIME
CREATE OR REPLACE TABLE workspace.projeto_datatran.gold_dim_time AS
SELECT DISTINCT
  data AS data_id,
  YEAR(data) AS ano,
  MONTH(data) AS mes,
  DAY(data) AS dia,
  DAYOFWEEK(data) AS dia_semana,
  quarter(data) AS trimestre
FROM workspace.projeto_datatran.datatran_silver
WHERE data IS NOT NULL
```

```sql
-- CREAT DIM LOCATION

CREATE OR REPLACE TABLE workspace.projeto_datatran.gold_dim_location AS
SELECT DISTINCT
  uf AS uf_id,
  municipio AS municipio_id,
  br AS br_id,
  km AS km_id,
  tipo_pista,
  sentido_via,
  uso_solo
FROM workspace.projeto_datatran.datatran_silver
WHERE uf IS NOT NULL AND municipio IS NOT NULL
```

```sql
-- CREAT DIM CONDITIONS

CREATE OR REPLACE TABLE workspace.projeto_datatran.gold_dim_conditions AS
SELECT DISTINCT
  condicao_metereologica AS condicao_metereologica_id,
  fase_dia AS fase_dia_id,
  tracado_via AS tracado_via_id
FROM workspace.projeto_datatran.datatran_silver
WHERE condicao_metereologica IS NOT NULL
```

```sql
--
CREATE OR REPLACE TABLE workspace.projeto_datatran.gold_fact_victim AS
SELECT DISTINCT
    pesid,
    id AS id_acidente,
    id_veiculo,
    estado_fisico,
    idade,
    sexo,
    tipo_envolvido
FROM workspace.projeto_datatran.datatran_silver
WHERE pesid IS NOT NULL
```
